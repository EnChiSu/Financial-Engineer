## Section 1:
### Demonstrate the function I create. You can simply execute and follow the instruction given.

```python
import pandas as pd
import math

def mortgage_return():
    # ask user to input parameters
    print("How much money you have to return?")
    m = int(input())
    print("How many years you have to return all the money?")
    y = int(input())
    print("What percentage is the interest rate per year? (give the number only)")
    r = float(input())/100
    
    # Create empty dataframe to store data
    ## (1)-1 produce row index for the dataframe
    period = list(range(y*12))
    period = [x+1 for x in period]
    period = ['第'+ str(x) +'期' for x in period]
    
    ## (1)-2 produce column index for the dataframe
    col = ['本 金(元)', '利 息(元)', '本金利息累計(元)']
    
    ## (2) Create the empty dataframe
    df2 = pd.DataFrame(index=period, columns=col)
    df2 = df2.fillna(0) # with 0s rather than NaNs
    
    # Calculate the principal needed to return in every period
    for i in list(range(y*12-1)):
        df2.iloc[i,0] = math.ceil(m/(12*y))
    df2.iloc[y*12-1,0] = m-sum(df2.iloc[:,0]) # Calculate the remaining principal need to return in the last period and store into the cell.

    
    # Calculate the interest you need to return in every period based on the interest rate you given, but before that we have to calculate the remaining mortgage you haven't returned in every period.
    ##(1) Calculate the remaining mortgage you haven't returned in every period.
    remaining = [m]*(y*12)
    for i in list(range(y*12-1)):
        remaining[i+1] = remaining[i]-df2.iloc[i,0]
        
    ##(2) Calculate the interest you need to return in every period   
    for i in list(range(y*12-1)):
        df2.iloc[i,1] = round(remaining[i]*(r/12),0)
    
    df2.iloc[y*12-1,1] = round(remaining[y*12-1]*(r/12),0) # Calculate the interest need to return in the last period and store into the last cell.

    
    # Calculate the accumulated return in every period
    ## (1) create empty list
    accumulate = [0]*(y*12)
    
    ## (2) calculate total return in every period (principal + interest in that period)
    for i in list(range(y*12)):
        accumulate[i] = df2.iloc[i,0] + df2.iloc[i,1]
    
    ## (3) add up total return from the first period to each period
    for i in list(range(1,y*12,1)):
        df2.iloc[i-1,2] = sum(accumulate[0:i])
    df2.iloc[y*12-1,2] = df2.iloc[y*12-2,2]+accumulate[y*12-1] # calculate the total return in the last period and store into the last cell
    
    # Create dataframe to present the parameters user input
    ## (1) produce row index for the dataframe
    row = ['本金', '期數(年)', '年利率', '平均每月攤還本金', '平均每月攤還利息', '全部利息']
    
    ## (2) Create the empty dataframe
    df1 = pd.DataFrame(index=row, columns=['Parameter'])
    df1 = df1.fillna(0)
    
    ## (3) Store the parameter into the dataframe
    df1.iloc[0] = m
    df1.iloc[1] = y
    df1.iloc[2] = r
    df1.iloc[3] = math.ceil(m/(12*y))
    df1.iloc[4] = '見參考下表'
    df1.iloc[5] = df2.iloc[y*12-1,2]-m
    
    # Demonstrate the result with two tables 
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)
    display(df1)
    display(df2)
```

```python
mortgage_return()
```

### Step 1:  (Build the second table)
 Ask users to input their mortgage, year to return, and interest rate respectively.

```python
print("How much money you have to return?")
m = int(input())
print("How many years you have to return all the money?")
y = int(input())
print("What is the interest rate per year? (give the number only)")
r = float(input())/100
```

### Step 2: 
 Create empty dataframe to store and demonstrate calculated data.

```python
# produce row index for the dataframe
period = list(range(y*12))
period = [x+1 for x in period]
period = ['第'+ str(x) +'期' for x in period]
period
```

```python
# produce column name for the dataframe
col = ['本 金(元)', '利 息(元)', '本金利息累計(元)']
```

```python
import pandas as pd
df2 = pd.DataFrame(index=period, columns=col)
df2 = df2.fillna(0) # with 0s rather than NaNs
df2
```

### Step 3: 
 Calculate the principal needed to return in every period, and store in the first column of the dataframe.

```python
# 將本金除以期數得到每期應還本金金額
import math
for i in list(range(y*12-1)):
    df2.iloc[i,0] = math.ceil(m/(12*y))
```

```python
#計算最後一期剩餘要還的本金，存入表格當中的最後一列 (第120期)
df2.iloc[y*12-1,0] = m-sum(df2.iloc[:,0])
df2
```

### Step 4: 
Calculate the interest needed to return in every period, and store in the second column of the dataframe.
<br>

Note:<br>
Before calculate the interest for every period, we have to calculate the remaining mortgage users haven't returned in every period.

```python
# Calculate the remaining mortgage you haven't returned in every period.remaining = [m]*(y*12)
for i in list(range(y*12-1)):
    remaining[i+1] = remaining[i]-df2.iloc[i,0]

remaining
```

```python
# Calculate interest for every period
for i in list(range(y*12-1)):
    df2.iloc[i,1] = round(remaining[i]*(r/12),0)
    
#計算最後一期剩餘要還的利息，存入表格當中的最後一欄
df2.iloc[y*12-1,1] = round(remaining[y*12-1]*(r/12),0)
df2
```

### Step 5: 
 Calculate the accumulated return in every period, and store in the third column of the dataframe.

```python
# create empty list
accumulate = [0]*(y*12)
```

```python
# store total return of every period (Principal + Interest)
for i in list(range(y*12)):
    accumulate[i] = df2.iloc[i,0] + df2.iloc[i,1]
```

```python
# add up the total return of each period from the first period and store into the dataframe
for i in list(range(1,y*12,1)):
    df2.iloc[i-1,2] = sum(accumulate[0:i])

df2.iloc[y*12-1,2] = df2.iloc[y*12-2,2]+accumulate[y*12-1]
df2
```

### Step 6: (Build the first table)
Create the parameter table to demonstrate users' inputs.

```python
import pandas as pd
# Create empty dataframe to store data for the parameters user input
## (1) produce row index for the dataframe
row = ['本金', '期數(年)', '年利率', '平均每月攤還本金', '平均每月攤還利息', '全部利息']
    
## (2) Create the empty dataframe
df1 = pd.DataFrame(index=row, columns=['Parameter'])
df1 = df1.fillna(0)
df1.iloc[0] = m
df1.iloc[1] = y
df1.iloc[2] = r
df1.iloc[3] = math.ceil(m/(12*y))
df1.iloc[4] = '見參考下表'
df1.iloc[5] = df2.iloc[y*12-1,2]-m
df1
```

## Section 3:
###  I have two learning lessons in this week's HW:

1. "accumulation" can't be added up by accumulation itself.
The obstacle I faced is during the **step 5** (see **Section 2**) I try to compute the third column *(本金利息累計(元))* through adding up accumulation itself from the first period. However, the accumulation should be calculated by adding up each period's "total" return, not each period's "accumulative" return. To solve the problem I create a new variable to record each period's "total" return. Using that variable I can add up from the first period to calculate the "accumulative" return for each period.<br>
(我遇到的問題是，我在做第二個表格的第三欄*(本金利息累計(元))*時(參見**Section 2**的**Step 5**)，原本試圖透過該欄位每一期從第一期加總到該期，但這個邏輯是有問題的，因為累積應該是每期"總額"的加總，而非每期"累積"的加總(如果是用每期"累積"的加總這樣算，前面期數的總額會被重複加很多次)，所以後來我另外設一個變項紀錄每期還的"總額"，也就是該期的應還本金加上應還利息，再依期數疊加上去就可以得到每期以還的累積總額)
